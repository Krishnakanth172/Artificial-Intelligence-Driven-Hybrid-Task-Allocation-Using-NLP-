```python
# Cell 2 - Imports and configuration
import random
import numpy as np
import pandas as pd
import re
from tqdm.auto import tqdm
import plotly.express as px
import plotly.graph_objects as go
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_recall_fscore_support
from sklearn.metrics.pairwise import cosine_similarity

RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# Quick config
SAMPLE_SIZE = 5000  # change this smaller (e.g., 5000) if runtime/memory is tight
NUM_EMPLOYEES = 50
NUM_AI_AGENTS = 20
EMBED_BATCH = 64
```

```python
# Cell 3 - Load the public HuggingFace Customer Support on Twitter dataset
# Dataset ID: "MohammadOthman/mo-customer-support-tweets-945k"
print("Loading dataset from Hugging Face... (may take 30-60s on first run)")
ds = load_dataset("MohammadOthman/mo-customer-support-tweets-945k", split="train")

print("Columns available:", ds.column_names)
# Heuristic to pick a textual column (robust)
candidate_text_cols = [c for c in ds.column_names if any(k in c.lower() for k in ("text","input","tweet","message","utter"))]
if len(candidate_text_cols) == 0:
    candidate_text_cols = [c for c in ds.column_names if ds.features[c].dtype == "string"]
text_col = candidate_text_cols[0]
print("Using text column:", text_col)

# Subsample for speed
n = min(SAMPLE_SIZE, len(ds))
ds_small = ds.shuffle(seed=RANDOM_SEED).select(range(n))
df = ds_small.to_pandas()
print("Sampled rows:", len(df))
df.head(3)
```

```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

# --- Text Cleaning Function ---
def clean_text(s):
    if not isinstance(s, str):
        return ""
    s = re.sub(r"http\S+", "", s)            # remove urls
    s = re.sub(r"@\w+", "", s)               # remove mentions
    s = re.sub(r"#", "", s)                  # drop hash sign
    s = re.sub(r"\s+", " ", s).strip()
    return s

# --- Apply cleaning ---
df["text_clean"] = df[text_col].astype(str).apply(clean_text)
df["len_chars"] = df["text_clean"].str.len().fillna(0)

# --- Basic stats ---
print("Text length: mean {:.1f}, median {}, max {}".format(
    df.len_chars.mean(), int(df.len_chars.median()), int(df.len_chars.max())
))

# --- Journal-quality Histogram ---
sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5), dpi=300)

sns.histplot(df["len_chars"], bins=50, color="navy", edgecolor="black", alpha=0.85)

plt.title("Distribution of Task Text Length (Characters)", fontsize=16, fontweight='bold', color='black')
plt.xlabel("Text Length (Characters)", fontsize=14, fontweight='bold', color='black')
plt.ylabel("Frequency", fontsize=14, fontweight='bold', color='black')

plt.xticks(fontsize=12, color='black')
plt.yticks(fontsize=12, color='black')

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(2.5)
    spine.set_color('black')

plt.tight_layout()
plt.savefig("text_length_distribution_300dpi.png", dpi=300, bbox_inches='tight')
plt.show()

# --- Show some random samples ---
df[["text_clean"]].sample(5, random_state=RANDOM_SEED).reset_index(drop=True)
```

```python
# --- Additional lightweight features ---
def extract_simple_feats(text):
    if not isinstance(text, str):
        return {
            "num_words": 0,
            "question_cnt": 0,
            "upper_ratio": 0.0,
            "has_url": 0,
            "has_mention": 0
        }
    words = text.split()
    num_words = len(words)
    question_cnt = text.count('?')
    letters = re.sub(r'[^A-Za-z]', '', text)
    upper_ratio = (sum(1 for c in letters if c.isupper()) / len(letters)) if len(letters) > 0 else 0.0
    has_url = int(bool(re.search(r'http\S+', text)))
    has_mention = int(bool(re.search(r'@\w+', text)))
    return {
        "num_words": num_words,
        "question_cnt": question_cnt,
        "upper_ratio": upper_ratio,
        "has_url": has_url,
        "has_mention": has_mention
    }

feat_dicts = df["text_clean"].apply(extract_simple_feats).tolist()
feats_df = pd.DataFrame(feat_dicts)
# merge
df = pd.concat([df.reset_index(drop=True), feats_df.reset_index(drop=True)], axis=1)

# quick view
df[["len_chars","num_words","question_cnt","upper_ratio","has_url","has_mention"]].describe().T
```

```python
# Cell 5 - Sentence-BERT embeddings for the tasks
print("Loading SentenceTransformer model (all-MiniLM-L6-v2).")
sbert_model = SentenceTransformer("all-MiniLM-L6-v2")

texts = df["text_clean"].tolist()

# compute embeddings in batches (returns np.array)
task_embeddings = sbert_model.encode(texts, batch_size=EMBED_BATCH, show_progress_bar=True, convert_to_numpy=True)
print("Embeddings shape:", task_embeddings.shape)
EMB_DIM = task_embeddings.shape[1]
```

```python
# Cell 6 - Simulate Employee & AI Agent metadata
# ----------------------------------------------
# We'll randomly generate profiles for humans and AI agents.

# --- Employees ---
employee_profiles = []
for i in range(NUM_EMPLOYEES):
    employee_profiles.append({
        "id": f"emp_{i}",
        "skill_level": np.random.randint(1, 6),    # 1 (low) - 5 (expert)
        "avg_workload": np.random.randint(1, 11),  # tasks per day
        "success_rate": np.random.uniform(0.6, 0.95)
    })

# --- AI Agents ---
ai_profiles = []
for j in range(NUM_AI_AGENTS):
    ai_profiles.append({
        "id": f"ai_{j}",
        "automation_capability": np.random.randint(1, 6),
        "latency": np.random.uniform(0.1, 2.0),  # seconds
        "success_rate": np.random.uniform(0.7, 0.99)
    })

# Combine into single assignee list
all_agents = employee_profiles + ai_profiles
agent_ids = [a["id"] for a in all_agents]
print(f"Generated {len(employee_profiles)} employees and {len(ai_profiles)} AI agents.")
```

```python
# Cell 7 - Create synthetic historical allocations
# -------------------------------------------------
# We'll simulate the "best" agent for each past task based on embedding similarity to skills.

# For simulation simplicity:
#  - If task length < 80 chars => more likely AI handles it
#  - If longer or complex => more likely human handles it

labels = []
for txt_len in df["len_chars"]:
    if txt_len < 80:
        if np.random.rand() < 0.7:
            labels.append(np.random.choice([a["id"] for a in ai_profiles]))
        else:
            labels.append(np.random.choice([e["id"] for e in employee_profiles]))
    else:
        if np.random.rand() < 0.7:
            labels.append(np.random.choice([e["id"] for e in employee_profiles]))
        else:
            labels.append(np.random.choice([a["id"] for a in ai_profiles]))

df["assigned_agent"] = labels
print(df["assigned_agent"].value_counts().head())
```

```python
from sklearn.preprocessing import StandardScaler

# Construct feature matrix: embeddings + engineered features
simple_feature_cols = ["len_chars","num_words","question_cnt","upper_ratio","has_url","has_mention"]
simple_feats = df[simple_feature_cols].fillna(0).to_numpy(dtype=float)

# Normalize simple features
scaler = StandardScaler()
simple_feats_scaled = scaler.fit_transform(simple_feats)

# Concatenate
X_all = np.hstack([task_embeddings, simple_feats_scaled])
print("Combined feature shape:", X_all.shape)  # should be (n, emb_dim + 6)

# Binary target as before
def is_ai(agent_id):
    return 1 if agent_id.startswith("ai_") else 0
y_binary = df["assigned_agent"].apply(is_ai).values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_all, y_binary, test_size=0.2, random_state=RANDOM_SEED, stratify=y_binary
)
print("Train size:", X_train.shape, "Test size:", X_test.shape)
```

```python
!pip install -q xgboost
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# --- RandomForest with class_weight='balanced' (no tuning) ---
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=RANDOM_SEED,
    n_jobs=-1,
    class_weight="balanced"
)
rf.fit(X_train, y_train)

# --- XGBoost tuned (smaller search, cv=2) ---
xgb = XGBClassifier(use_label_encoder=False, eval_metric="logloss", random_state=RANDOM_SEED, n_jobs=-1)

xgb_param_dist = {
    "n_estimators": randint(100, 300),
    "max_depth": randint(3, 8),
    "learning_rate": uniform(0.05, 0.2),
    "subsample": uniform(0.7, 0.3),
    "colsample_bytree": uniform(0.7, 0.3),
    "gamma": uniform(0, 3)
}

rs_xgb = RandomizedSearchCV(
    xgb, xgb_param_dist, n_iter=10, cv=2, scoring="f1", random_state=RANDOM_SEED, n_jobs=-1
)
rs_xgb.fit(X_train, y_train)
best_xgb = rs_xgb.best_estimator_
print("Best XGB params:", rs_xgb.best_params_)

# --- Ensemble Voting ---
ensemble = VotingClassifier(
    estimators=[("rf", rf), ("xgb", best_xgb)],
    voting="soft",
    n_jobs=-1
)
ensemble.fit(X_train, y_train)

clf = ensemble

# --- Evaluate ---
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=["Human", "AI"]))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```
